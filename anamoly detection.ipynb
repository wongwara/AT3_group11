```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
df = df.apply(LabelEncoder().fit_transform)
df.head()
```

```python
# Create a list of columns to use for creating the model
columns = [ 'lat', 'long',  'job', 'dob', 'acct_num', 'trans_num', 'unix_time', 'category', 'amt', 'is_fraud', 'merchant', 'merch_lat', 'merch_long']

# Create a new dataframe with only the columns we are interested in
df_model = df[columns]

```

```python
# Create a list of features to use for anomaly detection
features = [ 'lat', 'long','job', 'dob', 'acct_num', 'trans_num', 'unix_time', 'category', 'amt', 'is_fraud', 'merchant', 'merch_lat', 'merch_long']

# Create a new dataframe with only the features we are interested in
df_features = df[features]

```

```python
# Separate the dataframe into two parts: the training set and the test set
train_set = df_features[df_features['unix_time'] < 4100000]
test_set = df_features[df_features['unix_time'] >= 4100000]
print(test_set)
```

```python

from sklearn import preprocessing
  
for feature in features:
    if df_features[feature].dtype == 'string':
        le = preprocessing.LabelEncoder()
        df_features[feature] = le.fit_transform(df_features[feature])
```

```python
# Create a list of outliers
outliers = []

# For each feature, calculate the mean and standard deviation
for feature in features:
    mean = np.mean(train_set[feature])
    std = np.std(train_set[feature])
    
    # Find the rows in the test set that are outside the given threshold
    threshold = 3*std
    outliers.extend(test_set[test_set[feature] > mean + threshold].index)
    outliers.extend(test_set[test_set[feature] < mean - threshold].index)
    
# Create a new dataframe with only the customers with abnormal behaviors
df_outliers = df_model.loc[outliers]

# Print the dataframe
print(df_outliers)
```
